# Intro to ML Boot Camp - Day 4

## Decision Tree

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
bank = pd.read_csv('bank.csv')
bank.head()
```

### Pre-Processing

```python
bank.isnull().sum()
```

```python
bank = bank.drop(['contact', 'poutcome', 'day', 'month'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
columns = ['job', 'marital', 'education', 'default', 'housing', 'loan']

label_encoder = LabelEncoder()
target_encoder = LabelEncoder()

for col in columns:
    bank[col] = label_encoder.fit_transform(bank[col])
bank['deposit'] = target_encoder.fit_transform(bank['deposit'])
```

```python
bank.head(3)
```

```python
X = bank.iloc[:, :-1].values
y = bank.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Model

```python
from sklearn.tree import DecisionTreeClassifier
entropy = DecisionTreeClassifier(criterion='entropy', random_state=0)
gini = DecisionTreeClassifier(criterion='gini', random_state=0)
```

```python
entropy.fit(X_train, y_train)
gini_fitted = gini.fit(X_train, y_train)
```

```python
entropy_pred = entropy.predict(X_test)
gini_pred = gini.predict(X_test)
```

### Evaluation

```python
from sklearn.metrics import accuracy_score
entropy_score = accuracy_score(y_test, entropy_pred)
gini_score = accuracy_score(y_test, gini_pred)
```

```python
print('[entropy_score] {:.2f}'.format(entropy_score*100))
print('[gini_score] {:.2f}'.format(gini_score*100))
```

### Visualization

```python
from sklearn import tree
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8))
tree.plot_tree(gini_fitted)
plt.show()
```

## Random Forest (Bagging)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_20 = RandomForestClassifier(n_estimators=20, random_state=0)
rfc_50 = RandomForestClassifier(n_estimators=50, random_state=0)
rfc_100 = RandomForestClassifier(n_estimators=100, random_state=0)
```

```python
rfc_20.fit(X_train, y_train)
rfc_50.fit(X_train, y_train)
rfc_100.fit(X_train, y_train)
```

```python
rfc_20_pred = rfc_20.predict(X_test)
rfc_50_pred = rfc_50.predict(X_test)
rfc_100_pred = rfc_100.predict(X_test)
```

```python
print('Model accuracy with 20 estimators: {:.4f}'.format(accuracy_score(y_test, rfc_20_pred)*100))
print('Model accuracy with 50 estimators: {:.4f}'.format(accuracy_score(y_test, rfc_50_pred)*100))
print('Model accuracy with 100 estimators: {:.4f}'.format(accuracy_score(y_test, rfc_100_pred)*100))
```

## AdaBoost (Boosting)

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
adaboost_model = AdaBoostClassifier(base_estimator=gini, n_estimators=50, random_state=42)
```

```python
adaboost_model.fit(X_train, y_train)
```

```python
y_pred = adaboost_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

## XG Boost (Boosting)

```python
import xgboost as xgb
```

```python
xgb_model = xgb.XGBClassifier(base_estimator=gini, objective='binary:logistic', random_state=42)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(X_test)
```

```python
xgb_accuracy = accuracy_score(y_test, y_pred)
print("Accuracy (Binary Classification):", xgb_accuracy*100)
```
